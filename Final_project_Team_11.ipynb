```python
import time
from datetime import datetime, timedelta

# Dictionary to track last log times
last_seen = {}

def log_to_sheet(name):
    """Logs a face recognition event based on timing rules."""
    now = datetime.now()
    timestamp = now.strftime("%Y-%m-%d %H:%M:%S")

    # If unknown, log instantly
    if name == "Unknown":
        worksheet.append_row(["", name, timestamp, "Needs naming", timestamp])
        print(f"[LOGGED] Unknown face at {timestamp}")
        return

    # If known person: only log once every 5 minutes
    if name in last_seen:
        if now - last_seen[name] < timedelta(minutes=5):
            print(f"[SKIPPED] {name} recently logged, skipping...")
            return

    # Log known name
    worksheet.append_row(["", name, timestamp, "", timestamp])
    last_seen[name] = now
    print(f"[LOGGED] {name} at {timestamp}")
```

```python
import cv2
import os
import pickle
import mediapipe as mp

KNOWN_FACES_DIR = "known_faces"
ENCODINGS_FILE = "encodings.pkl"

# Use FaceMesh only
mp_mesh = mp.solutions.face_mesh
face_mesh = mp_mesh.FaceMesh(refine_landmarks=True)

def get_embedding(image):
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(rgb)

    if not results.multi_face_landmarks:
        return None

    embedding = []

    # Extract normalized landmark points
    for lm in results.multi_face_landmarks[0].landmark:
        embedding.extend([lm.x, lm.y, lm.z])

    return embedding


face_embeddings = []
labels = []

print("🔍 Processing known faces...")

for person_name in os.listdir(KNOWN_FACES_DIR):
    person_path = os.path.join(KNOWN_FACES_DIR, person_name)

    if not os.path.isdir(person_path):
        continue

    for filename in os.listdir(person_path):
        if filename.lower().endswith((".jpg", ".jpeg", ".png")):
            file_path = os.path.join(person_path, filename)
            image = cv2.imread(file_path)

            embedding = get_embedding(image)

            if embedding:
                face_embeddings.append(embedding)
                labels.append(person_name)
                print(f"✔ Processed: {person_name} -> {filename}")
            else:
                print(f"❌ No face detected: {filename}")

# Save encodings
with open(ENCODINGS_FILE, "wb") as f:
    pickle.dump({"embeddings": face_embeddings, "labels": labels}, f)

print("\n📁 Saved embeddings to:", ENCODINGS_FILE)
print("👤 Total faces encoded:", len(labels))
```

```python
import cv2
import numpy as np
import mediapipe as mp
from datetime import datetime
import gspread
from google.oauth2.service_account import Credentials
import pickle

# ------------------------ Google Sheet Setup ------------------------
SCOPES = [
    "https://www.googleapis.com/auth/spreadsheets",
    "https://www.googleapis.com/auth/drive"
]

creds = Credentials.from_service_account_file("face_recognition.json", scopes=SCOPES)
gc = gspread.authorize(creds)
sheet = gc.open("facerecognitionapp").worksheet("Sheet1")

# ----------------------- Load Face Encodings ------------------------
with open("encodings.pkl", "rb") as f:
    data = pickle.load(f)

known_embeddings = np.array(data["embeddings"])
known_labels = np.array(data["labels"])

# ---------------- Mediapipe Face Detection ----------------
mp_face_detection = mp.solutions.face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.6)

# ---------------- Smart Logging ----------------
last_seen = {}
LOG_INTERVAL = 30  # seconds

def log_to_sheet(name):
    now = datetime.now()
    sheet.append_row([
        sheet.row_count,
        name,
        now.strftime("%H:%M:%S"),
        "Detected",
        now.strftime("%d/%m/%Y")
    ])
    print(f"✔ Logged: {name} at {now.strftime('%H:%M:%S')}")


# ---------------- Webcam ----------------
cap = cv2.VideoCapture(0)

print("🎥 Webcam starting... press Q to stop")

while True:
    ret, frame = cap.read()
    if not ret:
        print("❌ Webcam not detected.")
        break

    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = mp_face_detection.process(rgb)

    if results.detections:
        for detection in results.detections:
            bbox = detection.location_data.relative_bounding_box
            h, w, _ = frame.shape
            x1, y1 = int(bbox.xmin * w), int(bbox.ymin * h)
            x2, y2 = x1 + int(bbox.width * w), y1 + int(bbox.height * h)

            face = rgb[y1:y2, x1:x2]

            if face.size == 0:
                continue

            embedding = np.mean(face, axis=(0,1))

            distances = np.linalg.norm(known_embeddings - embedding, axis=1)
            idx = np.argmin(distances)

            if distances[idx] < 50:
                name = known_labels[idx]
            else:
                name = "Unknown"

            # Display
            cv2.putText(frame, name, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0,255,0), 2)

            # Smart Logging
            now = datetime.now().timestamp()

            if name != "Unknown":
                if (name not in last_seen) or (now - last_seen[name] > LOG_INTERVAL):
                    log_to_sheet(name)
                last_seen[name] = now

    cv2.imshow("Face Recognition", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("🛑 Stopped by user")
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os
import time
import pickle
import face_recognition
import numpy as np
import gspread
from google.oauth2.service_account import Credentials
from datetime import datetime, timedelta

# ---------------- Google Sheets Setup ---------------- #
SCOPES = [
    "https://www.googleapis.com/auth/spreadsheets",
    "https://www.googleapis.com/auth/drive"
]
creds = Credentials.from_service_account_file("face_recognition.json", scopes=SCOPES)
gc = gspread.authorize(creds)
sheet = gc.open("facerecognitionapp").worksheet("Sheet1")
print("✔ Connected to Google Sheet!")

# ---------------- Load Known Faces ---------------- #
KNOWN_FACES_DIR = "known_faces"
ENCODINGS_FILE = "encodings.pkl"

if os.path.exists(ENCODINGS_FILE):
    with open(ENCODINGS_FILE, "rb") as f:
        known_faces = pickle.load(f)
else:
    known_faces = {}
    for person_name in os.listdir(KNOWN_FACES_DIR):
        person_dir = os.path.join(KNOWN_FACES_DIR, person_name)
        if not os.path.isdir(person_dir):
            continue
        for filename in os.listdir(person_dir):
            file_path = os.path.join(person_dir, filename)
            image = face_recognition.load_image_file(file_path)
            encs = face_recognition.face_encodings(image)
            if encs:
                known_faces[person_name] = encs[0]
    with open(ENCODINGS_FILE, "wb") as f:
        pickle.dump(known_faces, f)

print(f"👤 Total known faces: {len(known_faces)}")

# ---------------- Tracking Setup ---------------- #
SESSION_INTERVAL = timedelta(minutes=5)      # Log every 5 minutes
ABSENCE_THRESHOLD = timedelta(seconds=30)   # New session if absent > 30s

# Track last seen and last logged time
last_seen = {}
last_logged = {}

# ---------------- Start Webcam ---------------- #
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("❌ Webcam not detected.")
    exit()

print("🎥 Webcam starting... Press 'Q' to stop")

while True:
    ret, frame = cap.read()
    if not ret:
        continue

    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    face_locations = face_recognition.face_locations(rgb_frame)
    face_encodings = face_recognition.face_encodings(rgb_frame, face_locations)

    now = datetime.now()

    for face_encoding in face_encodings:
        matches = face_recognition.compare_faces(list(known_faces.values()), face_encoding)
        name = "Unknown"

        if True in matches:
            match_index = matches.index(True)
            name = list(known_faces.keys())[match_index]

        # Update last seen
        last_seen[name] = now

        # Check logging conditions
        if name not in last_logged or (now - last_logged[name]) >= SESSION_INTERVAL:
            sheet.append_row([name, now.strftime("%Y-%m-%d %H:%M:%S")])
            last_logged[name] = now
            print(f"Logged {name} at {now.strftime('%H:%M:%S')}")

    # Check for absences
    for person in list(last_seen.keys()):
        if now - last_seen[person] > ABSENCE_THRESHOLD:
            # Reset last_logged to allow new session if person returns
            last_logged.pop(person, None)
            last_seen.pop(person, None)

    # Show webcam frame
    cv2.imshow("Face Recognition", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
